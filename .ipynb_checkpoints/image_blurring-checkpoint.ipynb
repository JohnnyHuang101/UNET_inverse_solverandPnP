```python
import torch
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
def blur(image):
    
    transformer = transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((256,256)),
        transforms.ToTensor(),
    ])
    
    image_tensor = transformer(image).unsqueeze(0) #1,1,256,256
    _3x3_blur = torch.tensor([[1/16, 2/16, 1/16],
                              [2/16, 4/16, 2/16],
                              [1/16, 2/16, 1/16]])
    _3x3_tensor = _3x3_blur.view(1,1,3,3)
    
    
    blurred_tensor = F.conv2d(image_tensor,_3x3_tensor,padding=1)
    blurred_img = blurred_tensor.squeeze()
    blur_n_noise = blurred_img + (20/255)*torch.randn_like(blurred_img)
    
    blurred_img_np = blur_n_noise.numpy()

    # Plot the images
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1) 
    plt.imshow(image, cmap='gray')  
    plt.title('Original Image')  
    plt.axis('off') 
    
    plt.subplot(1, 2, 2) 
    plt.imshow(blurred_img_np, cmap='gray')
    plt.title('Blurred Image')
    plt.axis('off')
    plt.show()
```

```python
#example

test_image = Image.open("Train400/test_069.png")

blur(test_image)
```
